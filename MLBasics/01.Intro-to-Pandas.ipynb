#### Missing Data

```python
# Missing Data
# Pandas lists missing values as NaN

# Lets' Import car sales data with missing values
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# Let's Find the mean
car_sales_missing["Odometer"].mean()
```

```python
# Fill Odometer column missing values with mean
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean) # inplace is set to False by default 
```

```python
# Now let's check the original car_sales_missing DataFrame
car_sales_missing
```

```python
# Fill the Odometer missing values to the mean with inplace=True and, assign the values to the Odometer column.
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),
                                     inplace=True)
```

```python
car_sales_missing
```

We've filled some values but there's still missing values in car_sales_missing. Let's say we wanted to remove any rows which had missing data and only work with rows which had complete coverage. We can do this using .dropna().

```python
# Remove missing data
car_sales_missing.dropna()
```

It appears the rows with missing values have been removed, now let's check to make sure.

```python
car_sales_missing
```

Hmm, they're still there, can you guess why?

It's because .dropna() has inplace=False as default. We can either set inplace=True or reassign the car_sales_missing DataFrame.

```python
# The following two lines do the same thing

# Operation happens inplace without reassignment
car_sales_missing.dropna(inplace=True)

# car_sales_missing gets reassigned to same DataFrame but with dropped values  
car_sales_missing = car_sales_missing.dropna()
```

Now if check again, the rows with missing values are gone and the index numbers have been updated.

```python
car_sales_missing
```

```python
# save the dropped df with the dropped missing values to a new file
car_sales_missing.to_csv("car-sales-mssing-dropped.csv")
```

Instead of removing or filling data, what if you wanted to create it?

For example, creating a column called Seats for number of seats.

pandas allows for simple extra column creation on DataFrame's.

Three common ways are:

1. Adding a pandas.Series as a column.
2. Adding a Python list as a column.
3. By using existing columns to create a new column.

```python
# Create a column from a pandas Series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
# Fill lin the missing values for the Seats column with filna() function 
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Create a column from a Python list (list requires to have exact same lenght of rows)
engine_sizes = [1.3, 2.0, 3.0, 4.2, 1.6, 1, 2.0, 2.3, 2.0, 3.0]
car_sales["Engine Size"] = engine_sizes
car_sales
```

We can also make a column by directly combining the values of other columns. Such as, price per kilometre on the Odometer.

```python
# Column from other columns
car_sales["Price per KM"] = car_sales["Price"] / car_sales["Odometer (KM)"]
car_sales
```

```python
# Another column from Python List
fuel_economy = [7.5, 9.2, 5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
# Another column from other columns 
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

Let's say wewanted to shuffle the order of our DataFrame so we could split it into train, validation and test sets. And even though the order of our samples was random, we wanted to make sure.

To do so we could use .sample(frac=1).

.sample() randomly samples different rows from a DataFrame.

The frac parameter dictates the fraction, where 1 = 100% of rows, 0.5 = 50% of rows, 0.01 = 1% of rows.

We can also use .sample(n=1) where n is the number of rows to sample.

```python
# shuffle all data for a sample
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled 
```

Calling .reset_index() on a DataFrame resets the index numbers to their defaults. It also creates a new Index column by default which contains the previous index values. We can use the drop=True parameters bellow to drop it. 

```python
# Since the indexes are out of order lets reset the indexes of car_sales_sampled
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

Finally, what if we wanted to apply a function to a column. Such as, converting the Odometer column from kilometers to miles.

We can do so using the .apply() function and passing it a Python lambda function. We know there's about 1.6 kilometers in a mile, so if we  divide the value in the Odometer column by 1.6, it should convert it to miles.

```python
# Change the Odometer values from kilometres to miles
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
```
